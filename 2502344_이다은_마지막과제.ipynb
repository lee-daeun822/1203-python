```python
#과제39

def print_numbers(a):
  for i in range(1, a+1, 2):
    print(i, end=' ')
  print()

a = int(input("양수입력: "))
print_numbers(a)
```

```python
def numbers(a, b, c, d):
    n = [a, b, c, d]
    return max(n), min(n)

a = int(input("숫자1: "))
b = int(input("숫자2: "))
c = int(input("숫자3: "))
d = int(input("숫자4: "))

maximum, minimum = numbers(a, b, c, d)
print("최대값:", maximum)
print("최소값:", minimum)
```

```python
#43
def numbers(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

num = int(input("정수 입력: "))
print(numbers(num))
```

```python
#44
def numbers(a, b):
    for i in range(2, 10):
        for j in range(2, 10):
            if i == a and j == b:
                if i * j >= 30:
                    print(i * j)

a = int(input("숫자 입력: "))
b = int(input("숫자 입력: "))

numbers(a, b)
```
